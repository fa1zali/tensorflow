**Importing libraries**

```python
import tensorflow as tf
```

**Create Tensors**

```python
string = tf.Variable("this is a string", tf.string)
number = tf.Variable(123, tf.int16)
floating = tf.Variable(3.45, tf.float16)

print(string)
print(number)
print(floating)
```

Rank/Degree of Tensor

```python
rank1_tensor = tf.Variable(["this is a string"], tf.string)
print(rank1_tensor)

rank2_tensor = tf.Variable([["this is a string", "ok"], ["this is also a string","yes"]], tf.string)
print(rank2_tensor)
```

```python
tf.rank(rank1_tensor)
```

```python
tf.rank(rank2_tensor)
```

```python
tf.rank(number)
```

```python
tf.rank(tf.Variable([["Yes"],["No"]], tf.string))
```
